```python
import numpy as np
import pandas as pd

x = np.linspace(0, 10, 100)
y = np.sin(x)

my_array = np.array([x,y]).T

df = pd.DataFrame(my_array, columns = ['x','y'])
csv = df.to_csv()
```

```python
import IPython.display as ipython_display
bundle = {}
data = {}
data['type'] = "csv"
data['data'] = csv
bundle['application/vnd.perspective.arrow'] = data
ipython_display.display(bundle, raw=True)
```
